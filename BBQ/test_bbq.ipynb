```python
import pandas as pd
import json
```

```python
category = [
    "Age", 
    "Race_ethnicity", 
    # "Sexual_orientation", 
    # "Disability_status", 
    # "Race_x_SES", 
    "Gender_identity", 
    # "Race_x_gender", 
    # "Nationality", 
    "Religion", 
    "Physical_appearance", 
    # "SES"
]
```

```python
sampled_data = []

for cat in category:
    with open(f"data/{cat}.jsonl") as f:
        data = [json.loads(line) for line in f]
        sampled_data += data[:40]
```
